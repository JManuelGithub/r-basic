```python
import math
```

```python
def ecuaciongrado2(A,B,C):
    x1=(-B+math.sqrt(B**2-4*A*C))/(2*A)
    x2=(-B-math.sqrt(B**2-4*A*C))/(2*A)
    return (x1,x2)
ecuaciongrado2(1,2,1)
```

## Pregunta 2

```python
def isPalindrome(string):
    string=string.lower()
    string=string.replace(" ","")
    
    for i in range(0, int(len(string)/2)):
        if string[i] != string[len(string)-i-1]:
            return False
    return True
isPalindrome("La ruta nos aporto otro paso natural")
```

## Pregunta 4

```python
def cast2morse(s):
    s=s.upper()
    alpha=[chr(i) for i in range(ord('A'),ord('Z')+1)]
    alpha.append(" ")
    morse=['.-', '-...', '-.-,', '-..', 
           '.', '..-.', '--.', '....', '..', 
           '.---', '-.-', '.-..', '--', '-.', 
           '---', '.--.', '--.-', '.-.', '...',
           '-', '..-', '...-', '.--', '-..-', '-.--', '--..'," "]
    dic = dict(zip(alpha,morse))
    mensaje=[]
    for j in s:
        mensaje.append(dic[j])
    print(mensaje)
cast2morse("jose manuel")
```

## Pregunta 6

```python
def isimpar(x):
    for i in range(2,x):
        if (x%i==0):
            return False
    return True
isimpar(19)
```

## Pregunta 7

```python
def capital(s):
    return s.capitalize()
def capitalfrase(s):
    words=[]
    w=""
    l=s.split()
    for i in l:
        words.append(capital(i))
    for i in words:
        w+=i+" "
    w=w.rstrip()
    return w
capitalfrase("este asunto es importante")
```

## Pregunta 8

```python
def MCD(x,y):
    while(y):
        x, y = y, x % y
    return x
MCD(24,18)
```
